```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
import json
```

## Loading in data

```python
# read data in to a list
data_list = []
# open file and read data in to the array
# NOTE you're operating on `saved_data.txt` and not `data.txt`
with open("./saved_data.txt") as f:
    for line in f:
        # print(line) # we inserted newline separator after each message received in the server
        try:
            x = json.loads(line)
        except ValueError:
            print("{0} was not proper json".format(line))
        except:
            print("Unexpected Error")
        else:
            # https://www.tutorialspoint.com/python/python_exceptions.htm
            # print(x) # if you want to print the data side-by-side
            data_list.append(x)
```

```python
data = np.array(data_list)
```

```python
data.shape
```

```python
data[0]
```

## Cleaning data for specific article

```python
# NOW will work on data for a specific article
# ============================================

# will only work if first article seen was the "pandas" one
data[0]["pandas"]
```

```python
sample_data = np.array(data[0]["pandas"])
```

```python
sample_data
```

```python
sample_data.shape
```

```python
initial_time = sample_data[0][1]
```

```python
initial_time
```

```python
# subtract initial time from every element in 2nd column
# previously showed ms since epoch
# https://stackoverflow.com/questions/41946956/subtract-numpy-array-by-column
sample_data[...,1] -= initial_time
```

```python
sample_data
```

```python
plt.plot(sample_data)
```

```python
# convert integers to floats so that can divide later
sample_data = np.float_(sample_data)
```

```python
sample_data
```

```python
# divide all in column #2 by 1000 to convert ms to s
sample_data[:,1] *= 0.001
```

```python
# SINGLE ARTICLE now completely clean

"""
col idx0 of matrix is line number 
col idx1 is timestamp in s for when saw this line num
"""

sample_data
```

## Plotting data for specific article

```python
# PLOTTING data for that article
# ============================================
```

```python
# plot graph showing which line we look at as time passes
# as you can see, this data shows pretty linear reading

x = sample_data[:,1]  # col at idx1 represents how many seconds from start we saw specific line at
y = sample_data[:,0]  # what line we saw at that time
plt.plot(x,y)
plt.title('line numbers by time', fontsize=20)
plt.xlabel('time since start (s)')
plt.ylabel('line being looked at')
plt.show()
```
